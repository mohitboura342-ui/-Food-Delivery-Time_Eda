Exploratory Data Analysis is the first step in understanding the food delivery dataset before building a prediction model. The goal is to identify patterns, relationships, missing values, outliers, and factors that influence `Time_taken(min)`. [dergipark.org](https://dergipark.org.tr/en/pub/aita/issue/84471/1459560)
In this project, EDA helps us understand how distance, traffic, weather, driver details, and order timing affect delivery time. [github](https://github.com/Kaushik-Puttaswamy/Food-Delivery-Time-Prediction-Using-Machine-Learning)

## 2. Objective of EDA

The main objective of EDA is to discover which variables most strongly influence food delivery time and to prepare the dataset for regression modeling. [themoonlight](https://www.themoonlight.io/en/review/food-delivery-time-prediction-in-indian-cities-using-machine-learning-models)
EDA also helps check data quality, distribution of variables, and interaction effects between important features such as traffic and weather. [ieeexplore.ieee](https://ieeexplore.ieee.org/document/10962585/)

# Dataset Overview

This dataset is designed for **Food Delivery Time Prediction**, where the objective is to estimate the total delivery time of an order using delivery personnel information, restaurant and customer locations, environmental conditions, order details, and operational factors. It represents a real-world logistics problem commonly faced by online food delivery platforms such as Uber Eats, DoorDash, Zomato, Swiggy, and Deliveroo.

The dataset contains **20 columns**, including one target variable (**Time_taken(min)**) and nineteen input features describing delivery personnel, restaurant location, customer location, order timing, weather, traffic, vehicle characteristics, and city information. Since the target variable is continuous, this is a **supervised machine learning regression problem**.

# Target Variable

## Time_taken(min)

The **Time_taken(min)** column is the target variable of this dataset. It represents the **total time required to complete a food delivery**, measured in minutes. Since it is a continuous numerical value, predicting delivery time is formulated as a **regression problem** in supervised machine learning.

The target variable is influenced by multiple operational, geographical, environmental, and temporal factors, including travel distance, traffic density, weather conditions, restaurant preparation time, delivery personnel characteristics, vehicle type, multiple deliveries, festivals, and city conditions. Accurate prediction of this variable helps food delivery platforms provide reliable Estimated Time of Arrival (ETA), improve customer satisfaction, optimize delivery routes, reduce operational costs, and enhance logistics planning.

Mathematically, the prediction problem can be expressed as:

\[
\text{Time\_taken(min)} = f(X_1, X_2, X_3, ..., X_n)
\]

where:

- \(X_1\) = Delivery person characteristics
- \(X_2\) = Restaurant location
- \(X_3\) = Customer location
- \(X_4\) = Weather conditions
- \(X_5\) = Traffic density
- \(X_6\) = Vehicle information
- \(X_7\) = Order characteristics
- \(X_8\) = Festival status
- \(X_9\) = City type
- ...
- Output = **Time_taken(min)**

````python
df["Time_taken(min)"] = (
    df["Time_taken(min)"]
      .str.replace("(min) ", "", regex=False)
      .astype(int)
)
````

````python
df["Order_Date"] = pd.to_datetime(
    df["Order_Date"],
    format="%d-%m-%Y"
)
````

````python
df["Time_Orderd"] = pd.to_datetime(
    df["Time_Orderd"],
    format="%H:%M:%S",
    errors="coerce"
)

df["Time_Order_picked"] = pd.to_datetime(
    df["Time_Order_picked"],
    format="%H:%M:%S",
    errors="coerce"
)
````

````python
df["Preparation_Time"] = (
    df["Time_Order_picked"] -
    df["Time_Orderd"]
).dt.total_seconds()/60
````

````python
mask = df["Preparation_Time"] < 0
df.loc[mask,"Preparation_Time"] += 24*60
````
